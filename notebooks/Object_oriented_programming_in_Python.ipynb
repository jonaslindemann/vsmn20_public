## What is Object-Oriented Programming?

Object-oriented programming (OOP) is a programming paradigm that organizes code around **objects** rather than functions. The basic concept is to combine data and methods into a single entity called an object.

### Key Concepts:
- **Objects** often represent nouns (things): point, circle, car, student, bank account
- **Objects interact** with each other by sending messages (calling methods)
- **Methods** are verbs that define what the object can do: move(), draw(), deposit(), withdraw()
- **Attributes** (properties) describe the object's state and relationships to other objects
- **Classes** are blueprints that define how objects should be structured and behave

### Benefits of OOP:
- **Code reusability**: Write once, use many times
- **Modularity**: Easier to maintain and debug
- **Abstraction**: Hide complex implementation details
- **Encapsulation**: Protect data and control access
- **Inheritance**: Build on existing code
- **Polymorphism**: Use objects of different types interchangeably

# Function-Based Programming vs Object-Oriented Programming

Before diving into OOP, let's first look at how we might solve problems using a **function-based** (procedural) approach. This will help us understand why OOP can be more effective for certain types of problems.

## Example: Managing Points in 2D Space

Let's say we want to represent and manipulate points in a 2D coordinate system. Using a function-based approach, we might represent a point as a list and create separate functions to work with points:

````python
# Function-based approach to working with 2D points
# A point is represented as a list [x, y]

def create_point(x, y):
    """Create a new point with given coordinates"""
    return [x, y]

def move_point(point, dx, dy):
    """Move a point by dx and dy"""
    point[0] += dx
    point[1] += dy

def zero_point(point):
    """Reset a point to origin (0, 0)"""
    point[0] = 0.0
    point[1] = 0.0

def set_point(point, x, y):
    """Set point coordinates to specific values"""
    point[0] = x
    point[1] = y

def print_point(point):
    """Print point coordinates in a readable format"""
    print(f"Point: x = {point[0]}, y = {point[1]}")

def distance_from_origin(point):
    """Calculate distance from point to origin"""
    return (point[0]**2 + point[1]**2)**0.5
````

````python
# Create a point and display it
p = create_point(0.5, 0.0)
print_point(p)
print(f"Distance from origin: {distance_from_origin(p):.2f}")
````

````python
# Move the point and display the new position
move_point(p, 3.0, 2.0)
print("After moving by (3.0, 2.0):")
print_point(p)
print(f"Distance from origin: {distance_from_origin(p):.2f}")
````

````python
# Set the point to specific coordinates
set_point(p, -2.0, -1.0)
print("After setting to (-2.0, -1.0):")
print_point(p)
print(f"Distance from origin: {distance_from_origin(p):.2f}")
````

````python
# Let's create multiple points to see potential issues
p1 = create_point(1.0, 2.0)
p2 = create_point(-1.0, 3.0)

print("Point 1:")
print_point(p1)
print("Point 2:")
print_point(p2)

# What if we accidentally pass wrong arguments?
# This could cause problems:
# move_point(3.0, p1, 2.0)  # Wrong order of arguments!
````

## Problems with Function-Based Approach

While the function-based approach works, it has several limitations:

1. **No data protection**: Anyone can modify the internal structure `point[0] = "invalid"`
2. **Scattered functionality**: Point-related functions are spread throughout the code
3. **Parameter confusion**: Easy to mix up parameter order: `move_point(dx, point, dy)`
4. **No type checking**: Nothing prevents passing wrong data types
5. **Difficult to extend**: Adding new point types (3D, polar coordinates) requires rewriting functions
6. **Global namespace pollution**: Many function names in the global scope

**Object-oriented programming solves these problems by bundling data and related functions together!**

# Encapsulation and Data Protection

**Encapsulation** is one of the core principles of OOP. It means:
- **Bundling** data and methods together in a class
- **Controlling access** to internal data
- **Hiding implementation details** from users of the class

## Why is Encapsulation Important?

1. **Data Integrity**: Prevent invalid values from being assigned
2. **Interface Stability**: Internal changes don't break external code
3. **Debugging**: Easier to track where data is modified
4. **Security**: Sensitive data can be protected

## The Problem with Direct Attribute Access

In our current Point class, anyone can directly modify the attributes:

```python
p = Point()
p.x = "invalid"  # This should not be allowed!
p.y = [1, 2, 3]  # Neither should this!
```

Let's see how Python provides encapsulation mechanisms.

## Private Attributes in Python

Python uses **name mangling** to create private attributes:
- Add double underscore prefix: `__attribute_name`
- Python automatically renames these internally
- Cannot be accessed directly from outside the class

**Benefits of private attributes:**
- Prevent accidental modification
- Force users to use intended interface methods
- Allow validation and error checking
- Enable future changes without breaking external code

Let's create a Point class with private attributes and constructor parameters:

````python
class Point:
    """Point class with private attributes and parameterized constructor"""

    def __init__(self, x=0.0, y=0.0):
        """
        Initialize point with given coordinates

        Args:
            x (float): x-coordinate, defaults to 0.0
            y (float): y-coordinate, defaults to 0.0
        """
        self.__x = x  # Private attribute (name mangled)
        self.__y = y  # Private attribute (name mangled)

    def get_coordinates(self):
        """Return tuple of (x, y) coordinates"""
        return (self.__x, self.__y)
````

We can now create **Point** instances and assign the instance attributes at the same time:

````python
# Create Point objects with initial values
p1 = Point()           # Uses default values (0.0, 0.0)
p2 = Point(1.0, 2.0)   # Custom coordinates
p3 = Point(x=5.0)      # Only x specified, y defaults to 0.0

print("Point coordinates:")
print(f"p1: {p1.get_coordinates()}")
print(f"p2: {p2.get_coordinates()}")
print(f"p3: {p3.get_coordinates()}")
````

We can no longer assign the instance attributes directly anymore. Something goes wrong:

````python
# Try to access attributes that don't exist (will cause AttributeError)
p = Point(1.0, 2.0)

try:
    print("Trying to access p.x:")
    print(p.x)
except AttributeError as e:
    print(f"Error: {e}")

try:
    print("Trying to access p.y:")
    print(p.y)
except AttributeError as e:
    print(f"Error: {e}")

print("\nBut we can access coordinates through the proper method:")
print(f"Coordinates: {p.get_coordinates()}")
````

The reason for this is that the x and y properties are now private. The internal attributes **__ x** cannot be reached either:

````python
# Private attributes cannot be accessed directly
p = Point(3.0, 4.0)

try:
    print("Trying to access private attribute __x:")
    print(p.__x)
except AttributeError as e:
    print(f"Error: {e}")

# However, Python name mangling means it's still accessible (but you shouldn't do this!)
print(f"\nActual private attribute name: {dir(p)}")
print("Private attributes are name-mangled to _ClassName__attribute")
print(f"p._Point__x = {p._Point__x}")  # Don't do this in real code!
print(f"p._Point__y = {p._Point__y}")  # This breaks encapsulation!
````

We have now protected our internal instance variables from a user of the object (encapsulation). To access these, we must now add methods to assign and return values. This is often done with get / set methods in object-oriented programming. We add the set methods to be able to assign the instance variables.

````python
class Point:
    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    def x(self, value):
        self.__x = value

    def y(self, value):
        self.__y = value

    def set(self, x, y):
        self.__x = x
        self.__y = y
````

The class can now be created and used as in the code below:

````python
p = Point(12.0, 13.0)

p.set(2.0, 3.0)
p.x(42.0)
p.y(83.0)
````

To return values, we add the functions **x ()** and **y ()**.

````python
class Point:
    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    def x(self):
        return self.__x

    def set_x(self, value):
        self.__x = value

    def y(self):
        return self.__y

    def set_y(self, value):
        self.__y = value

    def set(self, x, y):
        self.__x = x
        self.__y = y


````

Now we have full access to the internal instance variables in the class:

````python
p = Point(12.0, 13.0)

p.set(2.0, 3.0)
p.set_x(42.0)
p.set_y(83.0)

print(p.x())
print(p.y())
````

## Python Properties - The Pythonic Way

**Properties** provide a clean, Pythonic way to implement getters and setters:

### Without Properties (Verbose):
```python
p.set_x(5.0)     # Setting value
print(p.get_x()) # Getting value
```

### With Properties (Clean):
```python
p.x = 5.0        # Setting value (looks like direct access)
print(p.x)       # Getting value (looks like direct access)
```

### How Properties Work:
1. **Define getter/setter methods** for attribute access
2. **Use `property()` function** to create property objects
3. **Access attributes naturally** while maintaining encapsulation
4. **Add validation/logging** in setter methods

### Benefits:
- **Clean syntax**: Looks like direct attribute access
- **Backwards compatibility**: Can add properties later without breaking code
- **Data validation**: Control what values are allowed
- **Computed properties**: Calculate values on-the-fly

Let's see properties in action:

````python
class Point:
    """Point class demonstrating properties with validation"""

    def __init__(self, x=0.0, y=0.0):
        # Use the property setters for validation during initialization
        self.x = x  # This calls the setter!
        self.y = y  # This calls the setter!

    def get_x(self):
        """Getter method for x coordinate"""
        print("Getting x value")
        return self.__x

    def set_x(self, value):
        """Setter method for x coordinate with validation"""
        print(f"Setting x to {value}")
        if not isinstance(value, (int, float)):
            raise TypeError("x coordinate must be a number")
        self.__x = float(value)

    def get_y(self):
        """Getter method for y coordinate"""
        print("Getting y value")
        return self.__y

    def set_y(self, value):
        """Setter method for y coordinate with validation"""
        print(f"Setting y to {value}")
        if not isinstance(value, (int, float)):
            raise TypeError("y coordinate must be a number")
        if value > 100.0:  # Example constraint
            print("Warning: y value is very large, capping at 100.0")
            self.__y = 100.0
        else:
            self.__y = float(value)

    # Create property objects that connect attribute access to methods
    x = property(get_x, set_x, doc="X coordinate property")
    y = property(get_y, set_y, doc="Y coordinate property")
````

It is now much easier to access the instance variables:

````python
p = Point()

p.x = 42.0
p.y = 84.0

print(p.x)
print(p.y)
````

Properties give us the protection of get / set methods, but with the simplicity of assigning instance variables directly. It also gives us the opportunity to postpone the decision to add get / set methods and to use direct access to instance variables until you need the added protection.
Protection.

## Python Properties Using Decorators - The Modern Way

**Decorators** provide an even cleaner way to create properties in Python. They are the preferred modern approach.

### `@property` decorator:
- Converts a method into a **read-only property**
- The method name becomes the property name
- Called automatically when accessing the attribute

### `@attribute.setter` decorator:
- Creates a **write-property** for an existing `@property`
- Must use the same method name as the property
- Called automatically when assigning to the attribute

### Advantages of Decorator Syntax:
- **Less code**: No need for separate `property()` calls
- **Cleaner**: Method definitions are grouped together
- **More readable**: Clear intent with decorators
- **Modern Python style**: Preferred in contemporary code

Here's our Point class rewritten using decorators:

````python
class Point:
    """Modern Point class using property decorators"""

    def __init__(self, x=0.0, y=0.0):
        """Initialize with validation through property setters"""
        self.x = x  # Calls the @x.setter method
        self.y = y  # Calls the @y.setter method

    @property
    def x(self):
        """Get x coordinate (read-only access)"""
        print("Reading x property")
        return self.__x

    @x.setter
    def x(self, value):
        """Set x coordinate with validation"""
        print(f"Setting x property to {value}")
        if not isinstance(value, (int, float)):
            raise TypeError(f"x must be a number, got {type(value)}")
        self.__x = float(value)

    @property
    def y(self):
        """Get y coordinate (read-only access)"""
        print("Reading y property")
        return self.__y

    @y.setter
    def y(self, value):
        """Set y coordinate with validation"""
        print(f"Setting y property to {value}")
        if not isinstance(value, (int, float)):
            raise TypeError(f"y must be a number, got {type(value)}")
        if value < -100 or value > 100:
            raise ValueError(f"y must be between -100 and 100, got {value}")
        self.__y = float(value)

    @property
    def distance_from_origin(self):
        """Computed property: distance from origin (read-only)"""
        return (self.__x**2 + self.__y**2)**0.5

    def __str__(self):
        """String representation of the point"""
        return f"Point({self.__x}, {self.__y})"
````

The class can be used in exactly the same way as before, but now we don't need the **property**-definitions at the end of the class.

````python
# Using properties with validation
p = Point()

print("=== Setting valid values ===")
p.x = 42.0
p.y = 84.0

print("\n=== Reading values ===")
print(f"x = {p.x}")
print(f"y = {p.y}")
print(f"Distance from origin = {p.distance_from_origin:.2f}")
print(f"Point: {p}")

print("\n=== Testing validation ===")
try:
    p.y = 150  # This will raise ValueError
except ValueError as e:
    print(f"Validation error: {e}")

try:
    p.x = "invalid"  # This will raise TypeError
except TypeError as e:
    print(f"Type error: {e}")
````

If you only want to implement a read-only property the @xxxx.setter decorator can be left out.

# Instance methods

*Main way of interacting with objects.* Due to the fact that all data is handled internally in the object, the methods can often be shortened.

A typical instance method in the **Point** class can be a method of moving the point a certain distance in the x and y directions.

````python
class Point:
    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    @property
    def x(self):
        print("@property of x")
        return self.__x

    @x.setter
    def x(self, x):
        print("@setter for x")
        self.__x = x

    @property
    def y(self):
        print("@property of y")
        return self.__y

    @y.setter
    def y(self, y):
        print("@setter for y")
        self.__y = y

    def set(self, x, y):
        self.__x = x
        self.__y = y

    def move(self, dx, dy): # <---- Ny move-metod.
        self.__x += dx
        self.__y += dy
````

We can now use **`.move(...)`** to move our points:

````python
p0 = Point()
p1 = Point()
p0.move(10.0, 20.0)
p1.move(-5.0, -5.0)

print(p0.x, p0.y)
print(p1.x, p1.y)
````

Another method that can be implemented is a method for copying the attributes from another instance of the same object **.copy_from()**:

````python
class Point:
    """Complete Point class with all methods"""

    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    @property
    def x(self):
        return self.__x

    @x.setter
    def x(self, x):
        self.__x = x

    @property
    def y(self):
        return self.__y

    @y.setter
    def y(self, y):
        self.__y = y

    def set(self, x, y):
        """Set both coordinates at once"""
        self.__x = x
        self.__y = y

    def move(self, dx, dy):
        """Move point by dx and dy"""
        self.__x += dx
        self.__y += dy

    def copy_from(self, p):
        """Copy coordinates from another point"""
        self.__x = p.x
        self.__y = p.y
````

How this method is used is shown below:

````python
# Demonstrating the copy_from method
p0 = Point()
p1 = Point()

p0.move(10.0, 20.0)
p1.move(-5.0, -5.0)

print("Before copy_from:")
print(f"p0: x={p0.x}, y={p0.y}")  # Fixed: was (p0.x, p0.x)
print(f"p1: x={p1.x}, y={p1.y}")

# Copy p0's coordinates to p1
p1.copy_from(p0)

print("\nAfter p1.copy_from(p0):")
print(f"p0: x={p0.x}, y={p0.y}")
print(f"p1: x={p1.x}, y={p1.y}")
print("Notice: p1 now has the same coordinates as p0")
````

## Special instance methods

There are a variety of special instance methods for implementing extra functionality on objects **`__init__(...)`** is one of them
**`__str__(...)`** is used to convert an object to a printable form with **`print()`**

Using **`print()`** on an instance without the **`__str__()`** method produces the following printout.

````python
p = Point()
print(p)
````

It would be nicer if the item could print in a more presentable way:

````python
class Point:
    def __init__(self, x=0.0, y=0.0):
        self.__x = x
        self.__y = y

    @property
    def x(self):
        return self.__x

    @x.setter
    def x(self, x):
        self.__x = x

    @property
    def y(self):
        return self.__y

    @y.setter
    def y(self, y):
        self.__y = y

    def set(self, x, y):
        self.__x = x
        self.__y = y

    def move(self, dx, dy):
        self.__x += dx
        self.__y += dy

    def copy_from(self, p):
        self.__x = p.x
        self.__y = p.y

    def __str__(self): # <---- New __str__ method
        return "Point("+str(self.__x)+", "+str(self.__y)+")"
````

If we use the same code to print, we get instead:

# Class Inheritance - Building on Existing Code

**Inheritance** allows you to create new classes based on existing ones, inheriting their attributes and methods while adding or modifying functionality.

## Key Concepts:

### Parent Class (Base/Super Class):
- The class being inherited from
- Provides common attributes and methods
- Example: `Point` class with x, y coordinates

### Child Class (Derived/Sub Class):
- The class that inherits from another class
- Gets all parent attributes and methods automatically
- Can add new features or override existing ones
- Example: `Circle` class inheriting from `Point`

## Inheritance Syntax:
```python
class ParentClass:
    # Parent implementation
    pass

class ChildClass(ParentClass):  # Inherit from ParentClass
    # Child implementation
    pass
```

## Benefits of Inheritance:
1. **Code Reuse**: Don't repeat common functionality
2. **Hierarchical Organization**: Natural relationships between classes
3. **Polymorphism**: Treat related objects uniformly
4. **Extensibility**: Add features without modifying existing code

## Real-world Example:
```
Vehicle (parent)
├── Car (inherits from Vehicle)
├── Motorcycle (inherits from Vehicle)
└── Truck (inherits from Vehicle)
```

Let's see inheritance in action with geometric shapes:

````python
class Circle(Point):  # Circle inherits from Point
    """
    Circle class that inherits from Point
    A circle is a point with a radius
    """

    def __init__(self, x=0.0, y=0.0, r=1.0):
        """
        Initialize a circle with center coordinates and radius

        Args:
            x (float): x-coordinate of center
            y (float): y-coordinate of center
            r (float): radius of the circle
        """
        # Call the parent class constructor to initialize x and y
        super().__init__(x, y)
        self.r = r  # Add radius property using the setter

    @property
    def r(self):
        """Get radius of the circle"""
        return self.__r

    @r.setter
    def r(self, value):
        """Set radius with validation"""
        if not isinstance(value, (int, float)):
            raise TypeError("Radius must be a number")
        if value <= 0:
            raise ValueError("Radius must be positive")
        self.__r = float(value)

    @property
    def area(self):
        """Calculate area of the circle (computed property)"""
        import math
        return math.pi * self.__r ** 2

    @property
    def circumference(self):
        """Calculate circumference of the circle (computed property)"""
        import math
        return 2 * math.pi * self.__r

    def copy_from(self, other_circle):
        """
        Copy data from another circle
        Overrides parent method to also copy radius
        """
        super().copy_from(other_circle)  # Copy x, y coordinates
        self.r = other_circle.r           # Copy radius

    def __str__(self):
        """String representation of the circle"""
        return f"Circle(center=({self.x}, {self.y}), radius={self.__r})"
````

## Using Inherited Classes

The `Circle` class now has:
- **All Point methods**: x, y properties, move(), distance_from_origin
- **Additional Circle methods**: r property, area, circumference
- **Overridden methods**: copy_from(), __str__()

Let's see inheritance in action:

````python
# Create a Point and a Circle
p = Point(1.0, 2.0)
c = Circle(2.0, 4.0, 3.0)

print("=== Initial objects ===")
print(f"Point: {p}")
print(f"Circle: {c}")

print("\n=== Circle-specific properties ===")
print(f"Circle radius: {c.r}")
print(f"Circle area: {c.area:.2f}")
print(f"Circle circumference: {c.circumference:.2f}")

print("\n=== Inherited functionality ===")
print(f"Circle center coordinates: x={c.x}, y={c.y}")
print(f"Point coordinates: x={p.x}, y={p.y}")

# Both Point and Circle have move() method (inherited)
print("\n=== Moving objects ===")
print("Before move:")
print(f"  Point: {p}")
print(f"  Circle: {c}")

p.move(1.0, 1.0)
c.move(1.0, 1.0)

print("After moving both objects by (1, 1):")
print(f"  Point: {p}")
print(f"  Circle: {c}")

print("\n=== Modifying circle properties ===")
c.r = 5.0
c.x = 0.0
c.y = 0.0
print(f"Modified circle: {c}")
print(f"New area: {c.area:.2f}")
````

# Composition - Objects Containing Other Objects

**Composition** is when objects contain references to other objects as part of their structure. Instead of inheriting behavior, objects are built by combining simpler objects.

## Composition vs Inheritance:

### Inheritance ("is-a" relationship):
- `Circle` **is a** `Point` (with added radius)
- Child class extends parent class
- Shares the same interface

### Composition ("has-a" relationship):
- `Line` **has** two `Point` objects (start and end points)
- `Car` **has** an `Engine` object
- `Rectangle` **has** four `Point` objects (corners)

## When to Use Composition:
- When objects are naturally made of other objects
- When you want more flexibility than inheritance provides
- When the relationship is "has-a" rather than "is-a"
- When you need to change behavior at runtime

## Benefits:
1. **Flexibility**: Can change components at runtime
2. **Loose Coupling**: Objects can be developed independently  
3. **Code Reuse**: Compose different combinations of objects
4. **Clear Relationships**: Models real-world "has-a" relationships

## Example: Line Class
A line connects two points, so it **contains** two Point objects:

````python
import math

class Line:
    """
    Line class demonstrating composition
    A line is composed of two Point objects
    """

    def __init__(self, p0=None, p1=None):
        """
        Initialize line with two points

        Args:
            p0: First point (start). If None, creates Point(0,0)
            p1: Second point (end). If None, creates Point(1,1)
        """
        # Composition: Line contains Point objects
        self.__p0 = p0 if p0 is not None else Point()
        self.__p1 = p1 if p1 is not None else Point(1.0, 1.0)

    @property
    def p0(self):
        """Get the first point of the line"""
        return self.__p0

    @p0.setter
    def p0(self, point):
        """Set the first point of the line"""
        if not isinstance(point, Point):
            raise TypeError("p0 must be a Point object")
        self.__p0 = point

    @property
    def p1(self):
        """Get the second point of the line"""
        return self.__p1

    @p1.setter
    def p1(self, point):
        """Set the second point of the line"""
        if not isinstance(point, Point):
            raise TypeError("p1 must be a Point object")
        self.__p1 = point

    @property
    def length(self):
        """Calculate length of the line (computed property)"""
        dx = self.__p1.x - self.__p0.x
        dy = self.__p1.y - self.__p0.y
        return math.sqrt(dx*dx + dy*dy)

    @property
    def midpoint(self):
        """Calculate midpoint of the line (computed property)"""
        mid_x = (self.__p0.x + self.__p1.x) / 2
        mid_y = (self.__p0.y + self.__p1.y) / 2
        return Point(mid_x, mid_y)

    def move(self, dx, dy):
        """Move the entire line by displacing both points"""
        self.__p0.move(dx, dy)
        self.__p1.move(dx, dy)  # Fixed: was (dy, dy)

    def __str__(self):
        """String representation of the line"""
        return f"Line from {self.__p0} to {self.__p1} (length: {self.length:.2f})"
````

Note that the **Line** class does not inherit from **Point**. It only refers to instances of type **Point**.

````python
# Create a line using default points
line = Line()
print(f"Default line: {line}")

# Modify points through properties
line.p0.x = 0.0
line.p0.y = 2.0
line.p1.x = 3.0
line.p1.y = 4.0

print(f"Modified line: {line}")
print(f"Line length: {line.length:.2f}")
print(f"Line midpoint: {line.midpoint}")

# Create line with custom points
start_point = Point(-1.0, -1.0)
end_point = Point(2.0, 3.0)
custom_line = Line(start_point, end_point)

print(f"\nCustom line: {custom_line}")

# Move the entire line
custom_line.move(1.0, 1.0)
print(f"After moving by (1,1): {custom_line}")
````

In this example, the **Point** instances cannot be modified as they can only be accessed by the **get_p0 ()** and **get_p1 ()** methods. The **x** and **y** attributes of the **Point** class can still be modified.

It would also have been possible to modify the class to allow the assignment of the instance properties **p0** and **p1**.

````python
class Line:
    def __init__(self):
        self.__p0 = Point()
        self.__p1 = Point()

    @property
    def p0(self):
        return self.__p0

    @p0.setter
    def p0(self, p):
        self.__p0 = p

    @property
    def p1(self):
        return self.__p1

    @p1.setter
    def p1(self, p):
        self.__p1 = p

    def __str__(self):
        return "Line from: " + str(self.__p0) + " to " + str(self.__p1)
````

The class can now be used as follows:

````python
line = Line()
line.p0.x = 0.0
line.p0.y = 2.0
line.p1.x = 3.0
line.p1.y = 4.0

print(line)

p3 = Point(5.0, 5.0)
p4 = Point(10.0, 10.0)
line.p0 = p3
line.p1 = p4

print(line)
````

A method for calculating the length can easily be added:

````python
import math

class Line:
    def __init__(self):
        self.__p0 = Point()
        self.__p1 = Point()

    @property
    def p0(self):
        return self.__p0

    @p0.setter
    def p0(self, p):
        self.__p0 = p

    @property
    def p1(self):
        return self.__p1

    @p1.setter
    def p1(self, p):
        self.__p1 = p

    @property
    def length(self):
        return math.sqrt(math.pow(self.p1.x - self.p0.x, 2) +
            math.pow(self.p1.y - self.p0.y, 2))

    def __str__(self):
        return "Line from: " + str(self.__p0) + " to " + str(self.__p1)
````

Examples of use:

````python
line = Line()
line.p0.x = 0.0
line.p0.y = 2.0
line.p1.x = 3.0
line.p1.y = 4.0

print(line.length)
````

# Polymorphism - One Interface, Many Forms

**Polymorphism** means "many forms" - it allows objects of different types to be treated uniformly through a common interface.

## Key Concepts:

### Dynamic Method Resolution:
- The **correct method is called** based on the actual object type
- Happens at **runtime**, not compile time
- If a method isn't found in the current class, Python looks in parent classes

### Duck Typing (Python's Approach):
> *"If it looks like a duck and quacks like a duck, it's probably a duck"*

- Python doesn't care about the object's type
- Only cares that the object has the required methods
- If an object has a `move()` method, it can be moved

## Benefits:
1. **Code Flexibility**: Work with objects without knowing their exact type
2. **Extensibility**: Add new types without changing existing code  
3. **Clean Interfaces**: Single way to interact with related objects
4. **Runtime Behavior**: Choose behavior based on actual object type

## Real-world Examples:
- All shapes can be `draw()`-n, regardless of whether they're circles, rectangles, or triangles
- All files can be `read()` from, whether they're text files, images, or videos
- All vehicles can `start()`, whether they're cars, motorcycles, or boats

Let's see polymorphism in action:

````python
# Polymorphism example: Different objects in the same list
shapes = []

# Add different types of objects to the list
shapes.append(Point(0.0, 1.0))           # Point object
shapes.append(Circle(2.0, 1.0, 3.0))     # Circle object
shapes.append(Line())                     # Line object
shapes.append(42.0)                       # float object
shapes.append(42)                         # int object
shapes.append("Hello")                    # string object

print("=== Polymorphism in action ===")
print("All objects in the list, regardless of type:")

# Polymorphic behavior: print() works on all objects
for i, shape in enumerate(shapes):
    print(f"{i}: {shape}")
    print(f"    Type: {type(shape).__name__}")

print("\n=== Every object has a __str__ method ===")
# This is why print() works - every object has __str__()
for shape in shapes:
    print(f"Calling __str__(): {shape.__str__()}")
````

What does print (42.0) really mean?

````python
print(42.0.__str__())
````

````python
# Polymorphism with geometric shapes that share common methods
geometric_shapes = []

geometric_shapes.append(Point(0.0, 1.0))
geometric_shapes.append(Circle(2.0, 1.0, 3.0))

print("=== Polymorphic method calls ===")
print("Both Point and Circle objects have move() method:")

for shape in geometric_shapes:
    print(f"Before move: {shape}")
    shape.move(5, 5)  # Polymorphic call - works on both types
    print(f"After move(5,5): {shape}")
    print()

print("=== Different behavior, same method name ===")
# Different classes can implement the same method differently
shapes_with_areas = [
    Circle(0, 0, 3),     # Circle has area = π * r²
    # We could add Rectangle, Triangle, etc. here
]

for shape in shapes_with_areas:
    if hasattr(shape, 'area'):  # Check if object has area property
        print(f"{shape}")
        print(f"  Area: {shape.area:.2f}")
    if hasattr(shape, 'circumference'):  # Check if object has circumference
        print(f"  Circumference: {shape.circumference:.2f}")
    print()
````

# 📚 Summary and Best Practices

## What We've Learned:

### 1. **Classes and Objects**
- Classes are blueprints, objects are instances
- `__init__()` method initializes objects
- `self` refers to the current instance

### 2. **Encapsulation**
- Use private attributes with `__` prefix
- Control access with properties (`@property`, `@setter`)
- Validate data in setter methods

### 3. **Inheritance**
- Child classes inherit from parent classes: `class Child(Parent):`
- Use `super()` to call parent methods
- Override methods to customize behavior

### 4. **Composition**
- Objects can contain other objects
- Use when you have "has-a" relationships
- More flexible than inheritance

### 5. **Polymorphism**
- Different objects can share the same interface
- Python uses duck typing
- Write code that works with multiple object types

## Python OOP Best Practices:

### Class Design:
- **Use descriptive class names**: `BankAccount` not `BA`
- **Follow PascalCase**: `MyClass`, not `my_class`
- **Add docstrings**: Document what the class does
- **Keep classes focused**: Single responsibility principle

### Method Design:
- **Use snake_case for methods**: `calculate_area()`
- **Methods should do one thing**: Keep them simple
- **Use `@property` for computed values**: Don't store what you can calculate
- **Validate inputs**: Check types and ranges in setters

### Inheritance Guidelines:
- **Use inheritance for "is-a" relationships**: `Circle` is a `Shape`
- **Use composition for "has-a" relationships**: `Car` has an `Engine`
- **Prefer composition over inheritance**: More flexible
- **Keep inheritance hierarchies shallow**: Max 3-4 levels

### Code Organization:
- **Import at the top**: Keep imports organized
- **One class per file** (for larger projects)
- **Use meaningful variable names**: `radius` not `r`
- **Add type hints** (Python 3.5+): `def __init__(self, x: float, y: float)`
