# 3. Data Validation and Error Handling

Scientific computing requires robust error handling and data validation to ensure reliable results. This should be standard practice for all plotting workflows.

## Essential Data Validation

````python
def create_scientific_plot(x_data, y_data, title="Scientific Plot", **kwargs):
    """
    Create a publication-quality plot with comprehensive error checking.

    Parameters:
    -----------
    x_data, y_data : array-like
        Input data arrays
    title : str
        Plot title
    **kwargs : dict
        Additional plotting parameters

    Returns:
    --------
    fig, ax : matplotlib figure and axis objects
    """

    # Data validation
    x_data = np.asarray(x_data)
    y_data = np.asarray(y_data)

    # Check for common issues
    if len(x_data) != len(y_data):
        raise ValueError(f"Array length mismatch: x={len(x_data)}, y={len(y_data)}")

    if len(x_data) == 0:
        raise ValueError("Empty data arrays provided")

    if np.any(np.isnan(x_data)) or np.any(np.isnan(y_data)):
        print("Warning: NaN values detected in data")
        # Handle NaN values
        valid_mask = ~(np.isnan(x_data) | np.isnan(y_data))
        x_data = x_data[valid_mask]
        y_data = y_data[valid_mask]
        print(f"   Removed {np.sum(~valid_mask)} invalid points")

    # Create plot with default scientific formatting
    fig, ax = plt.subplots(figsize=(10, 6))

    # Apply default scientific styling
    plot_kwargs = {
        'linewidth': 2,
        'alpha': 0.8,
        'color': 'blue'
    }
    plot_kwargs.update(kwargs)

    ax.plot(x_data, y_data, **plot_kwargs)
    ax.set_title(title, fontsize=14)
    ax.grid(True, alpha=0.3)

    # Data summary
    print(f"Plot created successfully:")
    print(f"   Data points: {len(x_data)}")
    print(f"   X range: [{x_data.min():.3f}, {x_data.max():.3f}]")
    print(f"   Y range: [{y_data.min():.3f}, {y_data.max():.3f}]")

    return fig, ax
````

````python
import numpy as np

x = np.linspace(-2.0, 2.0, 100)
y = x**3

fig, ax = create_scientific_plot(x, y)
````

# 4. Basic Plotting and Customization

## Scientific Data with NumPy Arrays

NumPy arrays are the foundation of scientific computing in Python. Matplotlib is designed to work seamlessly with NumPy, enabling efficient plotting of mathematical functions, experimental data, and simulation results.

### Creating Coordinate Arrays

Create coordinate arrays for function plotting. np.linspace() is preferred over np.arange() for scientific computing because it guarantees exact endpoint inclusion and avoids floating-point errors.

````python
x = np.linspace(0.0, 2*np.pi, 50)  # 50 points from 0 to 2π (inclusive)
y = np.sin(x)                      # Vectorized computation - all points at once

# Scientific computing tip: Always verify your arrays before plotting
print(f"Function properties:")
print(f"Min/Max x: [{x.min():.3f}, {x.max():.3f}]")
print(f"Min/Max y: [{y.min():.3f}, {y.max():.3f}]")
````

### Professional Plot Creation

````python
# Create a scientific-quality plot with best practices

fig, ax = plt.subplots(figsize=(10, 6))  # Object-oriented approach recommended

# Plot with clear styling

ax.plot(x, y, 'b-', linewidth=2, label='sin(x)', alpha=0.8)
ax.set_xlabel('x (radians)', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('Sine Function: Mathematical Model Visualization', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)

# Add scientific context: mark important points

ax.axhline(y=0, color='k', linestyle='--', alpha=0.5, linewidth=1)  # y=0 reference
ax.axvline(x=np.pi/2, color='r', linestyle=':', alpha=0.5, label='π/2')
ax.axvline(x=3*np.pi/2, color='r', linestyle=':', alpha=0.5, label='3π/2')

# Set axis limits for better visualization

ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)

# Add π-based x-axis labels for scientific clarity

pi_ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
pi_labels = ['0', 'π/2', 'π', '3π/2', '2π']
ax.set_xticks(pi_ticks)
ax.set_xticklabels(pi_labels)

plt.tight_layout()
````

Generate data for comparison: sine and cosine functions.

## Single-Panel Publication Figure

Example: Publication-quality single-panel figure.
Generate realistic scientific data: enzyme kinetics.

````python
substrate_conc = np.logspace(-2, 1, 20)  # Substrate concentration (mM)
vmax = 100                               # Maximum velocity (μmol/min/mg)
km = 0.5                                 # Michaelis constant (mM)

# Michaelis-Menten equation

velocity_theory = (vmax * substrate_conc) / (km + substrate_conc)

# Add experimental noise (realistic for enzyme assays)

np.random.seed(42)
velocity_exp = velocity_theory + np.random.normal(0, 5, len(substrate_conc))

# Create publication-quality figure
fig, ax = plt.subplots(figsize=(6.4, 4.8))  # Single-column width

# Plot experimental data

ax.scatter(substrate_conc, velocity_exp,
           color='#1f77b4', s=40, alpha=0.7,
           edgecolors='white', linewidth=0.5,
           label='Experimental data', zorder=3)

# Plot theoretical curve

substrate_smooth = np.logspace(-2, 1, 200)
velocity_smooth = (vmax * substrate_smooth) / (km + substrate_smooth)
ax.plot(substrate_smooth, velocity_smooth,
        color='#d62728', linewidth=2,
        label='Michaelis-Menten model', zorder=2)

# Professional formatting

ax.set_xlabel('Substrate concentration (mM)', fontweight='bold')
ax.set_ylabel('Velocity (μmol/min/mg protein)', fontweight='bold')
ax.set_title('Enzyme Kinetics Analysis', fontweight='bold', pad=20)

# Use log scale for x-axis (common in biochemistry)

ax.set_xscale('log')
ax.set_xlim(0.01, 10)
ax.set_ylim(0, 110)

# Professional legend

ax.legend(loc='lower right', frameon=True, fancybox=False,
          shadow=False, framealpha=0.9)

# Add statistical information (common in scientific figures)

r_squared = np.corrcoef(velocity_theory, velocity_exp)[0, 1]**2
ax.text(0.05, 0.95, f'R² = {r_squared:.3f}',
        transform=ax.transAxes, fontsize=9,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

plt.tight_layout()
````

````python
categories = ['Method A', 'Method B', 'Method C', 'Method D']
values = [23, 45, 56, 78]
errors = [3, 5, 4, 6]

fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(categories, values, yerr=errors, capsize=5,
              color='lightcoral', edgecolor='darkred', linewidth=1,
              alpha=0.7)

# Add value labels on bars
for bar, value in zip(bars, values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{value}', ha='center', va='bottom', fontweight='bold')

ax.set_ylabel('Performance Metric')
ax.set_title('Method Comparison with Error Bars')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
````

## Exercise 1: Climate Data Analysis

**Scientific Context**: Global temperature anomalies provide critical evidence for climate change research. We'll analyze historical temperature data to identify trends and create publication-quality visualizations.

**Research Question**: How have global temperature anomalies changed over the past century, and can we identify significant warming trends?

````python
# Generate realistic climate data (based on actual temperature patterns)
np.random.seed(42)
years = np.arange(1880, 2024)  # Historical period similar to real datasets
n_years = len(years)

# Create realistic temperature anomaly pattern
base_trend = np.zeros(n_years)
base_trend[:100] = 0.02 * (years[:100] - 1880)  # Slow warming 1880-1980
base_trend[100:] = base_trend[99] + 0.08 * (years[100:] - 1980)  # Accelerated warming

# Add realistic variability
from scipy.ndimage import gaussian_filter1d
enso_cycle = 0.3 * np.sin(2 * np.pi * years / 4.5) * np.random.rand(n_years)
volcanic_events = np.random.choice([0, -0.8], size=n_years, p=[0.97, 0.03])
annual_noise = np.random.normal(0, 0.2, n_years)

# Combine all components
temp_anomaly = base_trend + enso_cycle + volcanic_events + annual_noise
temp_anomaly = gaussian_filter1d(temp_anomaly, sigma=0.5)

# Add realistic uncertainty estimates
uncertainty = 0.15 - 0.1 * (years - 1880) / (2023 - 1880)
uncertainty = np.maximum(uncertainty, 0.05)
measured_temp = temp_anomaly + np.random.normal(0, uncertainty)

# Statistical analysis
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(years, measured_temp)
trend_line = slope * years + intercept

# Create publication-quality climate visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Panel A: Time series with uncertainty
ax1.fill_between(years, measured_temp - uncertainty, measured_temp + uncertainty,
                alpha=0.3, color='lightblue', label='Uncertainty (±1σ)')

ax1.plot(years, measured_temp, 'o-', color='steelblue', markersize=1.5,
         linewidth=0.8, alpha=0.7, label='Annual temperature')

# Moving average
window = 10
moving_avg = np.convolve(measured_temp, np.ones(window)/window, mode='valid')
moving_years = years[window//2:-window//2+1]
ax1.plot(moving_years, moving_avg, color='red', linewidth=3,
         label=f'{window}-year moving average')

ax1.plot(years, trend_line, '--', color='darkred', linewidth=2,
         label=f'Linear trend ({slope*100:.2f}°C/decade)')

ax1.set_ylabel('Temperature Anomaly (°C)')
ax1.set_title('A. Global Temperature Anomalies (1880-2023)',
              fontweight='bold', loc='left')
ax1.legend(loc='upper left', fontsize=9)
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3, linewidth=1)

# Panel B: Decadal analysis
decades = np.arange(1880, 2030, 10)
decade_means = []
decade_stds = []

for decade in decades[:-1]:
    decade_mask = (years >= decade) & (years < decade + 10)
    if np.any(decade_mask):
        decade_means.append(measured_temp[decade_mask].mean())
        decade_stds.append(measured_temp[decade_mask].std())
    else:
        decade_means.append(np.nan)
        decade_stds.append(np.nan)

decade_centers = decades[:-1] + 5
valid_mask = ~np.isnan(decade_means)

bars = ax2.bar(decade_centers[valid_mask], np.array(decade_means)[valid_mask],
               width=8, alpha=0.7, color='coral', edgecolor='darkred', linewidth=1)

# Color bars by temperature (cooler = blue, warmer = red)
for bar, temp in zip(bars, np.array(decade_means)[valid_mask]):
    if temp < 0:
        bar.set_color('lightblue')
        bar.set_edgecolor('blue')
    else:
        intensity = min(temp / 0.8, 1.0)
        bar.set_color(plt.cm.Reds(0.4 + 0.6 * intensity))

ax2.set_xlabel('Decade')
ax2.set_ylabel('Mean Temperature Anomaly (°C)')
ax2.set_title('B. Decadal Temperature Averages', fontweight='bold', loc='left')
ax2.grid(True, alpha=0.3, axis='y')
ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3, linewidth=1)

plt.tight_layout()
````

## Exercise 2: Gene Expression Analysis

**Scientific Context**: Gene expression profiling is fundamental to understanding cellular responses to treatments, diseases, or developmental stages. We'll analyze RNA-seq data to identify differentially expressed genes and create volcano plots.

**Research Question**: Which genes are significantly upregulated or downregulated in response to a drug treatment, and how can we visualize the statistical significance vs. effect size?

````python
# Generate realistic RNA-seq differential expression data
import pandas as pd
from scipy.stats import false_discovery_control

np.random.seed(42)
n_genes = 15000  # Typical number of detected genes in RNA-seq

# Create gene names
gene_names = [f'Gene_{i:05d}' for i in range(n_genes)]

# Simulate log2 fold-changes with realistic distribution
log2fc = np.random.normal(0, 0.5, n_genes)  # Most genes near zero change

# Add some truly differentially expressed genes
n_de_genes = int(0.15 * n_genes)  # 15% differentially expressed
de_indices = np.random.choice(n_genes, n_de_genes, replace=False)

# Strongly upregulated genes
n_up = n_de_genes // 3
up_indices = de_indices[:n_up]
log2fc[up_indices] = np.random.normal(2.5, 0.8, n_up)

# Strongly downregulated genes
n_down = n_de_genes // 3
down_indices = de_indices[n_up:n_up+n_down]
log2fc[down_indices] = np.random.normal(-2.2, 0.7, n_down)

# Simulate p-values
p_values = np.random.uniform(0, 1, n_genes)
p_values[de_indices] = np.random.beta(0.1, 10, len(de_indices))

# Multiple testing correction
adj_p_values = false_discovery_control(p_values, method='bh')

# Create dataset
gene_data = pd.DataFrame({
    'gene_name': gene_names,
    'log2fc': log2fc,
    'p_value': p_values,
    'adj_p_value': adj_p_values
})

# Define significance thresholds
fc_threshold = 1.0  # |log2FC| > 1 (2-fold change)
p_threshold = 0.05  # Adjusted p-value < 0.05

# Classify genes
gene_data['significant'] = (np.abs(gene_data['log2fc']) > fc_threshold) & (gene_data['adj_p_value'] < p_threshold)
gene_data['upregulated'] = (gene_data['log2fc'] > fc_threshold) & (gene_data['adj_p_value'] < p_threshold)
gene_data['downregulated'] = (gene_data['log2fc'] < -fc_threshold) & (gene_data['adj_p_value'] < p_threshold)

# Create publication-quality genomics visualization
fig = plt.figure(figsize=(14, 10))

# Panel A: Volcano Plot
ax1 = plt.subplot(2, 2, 1)

# Plot non-significant genes
ns_genes = gene_data[~gene_data['significant']]
ax1.scatter(ns_genes['log2fc'], -np.log10(ns_genes['adj_p_value']),
           c='lightgray', alpha=0.5, s=10, label='Not significant')

# Plot significantly upregulated genes
up_genes = gene_data[gene_data['upregulated']]
ax1.scatter(up_genes['log2fc'], -np.log10(up_genes['adj_p_value']),
           c='red', alpha=0.7, s=15, label=f'Upregulated ({len(up_genes)})')

# Plot significantly downregulated genes
down_genes = gene_data[gene_data['downregulated']]
ax1.scatter(down_genes['log2fc'], -np.log10(down_genes['adj_p_value']),
           c='blue', alpha=0.7, s=15, label=f'Downregulated ({len(down_genes)})')

# Add threshold lines
ax1.axvline(x=fc_threshold, color='black', linestyle='--', alpha=0.5)
ax1.axvline(x=-fc_threshold, color='black', linestyle='--', alpha=0.5)
ax1.axhline(y=-np.log10(p_threshold), color='black', linestyle='--', alpha=0.5)

ax1.set_xlabel('log₂ Fold Change')
ax1.set_ylabel('-log₁₀ Adjusted P-value')
ax1.set_title('A. Volcano Plot', fontweight='bold', loc='left')
ax1.legend(loc='upper right', fontsize=8)
ax1.grid(True, alpha=0.3)

# Panel B: Fold Change Distribution
ax2 = plt.subplot(2, 2, 2)
ax2.hist(gene_data['log2fc'], bins=50, alpha=0.7, color='skyblue',
         edgecolor='black', linewidth=0.5)
ax2.axvline(x=fc_threshold, color='red', linestyle='--', linewidth=2,
           label=f'FC threshold (±{fc_threshold})')
ax2.axvline(x=-fc_threshold, color='red', linestyle='--', linewidth=2)
ax2.axvline(x=0, color='black', linestyle='-', alpha=0.5)
ax2.set_xlabel('log₂ Fold Change')
ax2.set_ylabel('Number of Genes')
ax2.set_title('B. Fold Change Distribution', fontweight='bold', loc='left')
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

# Panel C: P-value Distribution
ax3 = plt.subplot(2, 2, 3)
ax3.hist(gene_data['p_value'], bins=50, alpha=0.7, color='lightcoral',
         edgecolor='black', linewidth=0.5)
ax3.axvline(x=p_threshold, color='blue', linestyle='--', linewidth=2,
           label=f'Significance threshold ({p_threshold})')
uniform_height = len(gene_data) / 50
ax3.axhline(y=uniform_height, color='green', linestyle='-', linewidth=2,
           label='Expected (uniform)', alpha=0.7)
ax3.set_xlabel('Raw P-value')
ax3.set_ylabel('Number of Genes')
ax3.set_title('C. P-value Distribution', fontweight='bold', loc='left')
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Panel D: Summary Statistics
ax4 = plt.subplot(2, 2, 4)
categories = ['Total\nGenes', 'Upregulated', 'Downregulated', 'Non-significant']
counts = [len(gene_data), len(up_genes), len(down_genes), len(ns_genes)]
colors = ['gray', 'red', 'blue', 'lightgray']

bars = ax4.bar(categories, counts, color=colors, alpha=0.7,
               edgecolor='black', linewidth=1)

# Add count labels
for bar, count in zip(bars, counts):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + height*0.02,
             f'{count:,}', ha='center', va='bottom', fontweight='bold')

ax4.set_ylabel('Number of Genes')
ax4.set_title('D. Summary Statistics', fontweight='bold', loc='left')
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
````
